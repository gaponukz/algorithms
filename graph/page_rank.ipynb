```python
import networkx
import matplotlib.pyplot as plt
```

```python
graph = networkx.gnp_random_graph(21, 0.6, directed=True)
networkx.draw(graph, with_labels=True, font_color='black', font_size=7, node_color='yellow')
```

```python
pages_importance = {page: 0 for page in graph.nodes()}
print(pages_importance)
```

```python
links_from = {page: len(list(graph.neighbors(page))) for page in graph.nodes()}
links_to = {page: 0 for page in graph.nodes()}

print(links_from)
```

```python
for page in graph.nodes():
    for item in graph.neighbors(page):
        links_to[item] += 1

print(links_to)
```

```python
for page in graph.nodes():
    pages_importance[page] = links_to[page] / links_from[page]
```

```python
print(max(pages_importance, key=pages_importance.get))
```
